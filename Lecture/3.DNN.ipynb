```python
import tensorflow as tf
from tensorflow import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
#x_train, x_test 이미지 데이터 (입력 데이터)
#y_train, y_test 이미지에 대한 레이블(정답)
x_train = x_train / 255. #훈련데이터 픽셀값을 0과1사이로 정규화

print("STEP1 데이터 불러오기")
print("x_train: ", x_train)
print("y_train: ", y_train)



#Layer Design
f_layer = keras.layers.Flatten(input_shape=(28, 28, 1),name='flatten') #2차원이상 다차원데이터를 1차원으로 바꿔줌
#Flatten : 데이터 전처리 DNN은 입력데이터가 1차원이어야되는데 그 작업을 해줌

h_layer = keras.layers.Dense(units=100, activation='relu',name='hidden_layer') #노드수 100개 sigmoid -> relu 변경해서 테스트

o_layer = keras.layers.Dense(units=10, activation='softmax',name='output_layer')
# Dense - 레이어 만들어줌 / units = 노드의 수 / activation (시그노이드함수를 softmax로 지정한것)

print("STEP2 Flatten 2차원백터를 1차원으로, Dense 연결층생성")
print("f_layer: ", f_layer)
print("h_layer: ", h_layer)
print("o_layer: ", o_layer)


#Model build
model = keras.Sequential([f_layer, h_layer, o_layer]) #모델 생성 / 나중에 가운데에 히든레이어도 넣으면됨
print("model.summary: ")
model.summary() #모델 구조 요약해서 출력



#Model learning method setting
model.compile(loss='sparse_categorical_crossentropy', optimizer='SGD', metrics=['accuracy']) #학습 방법을 정해줌
#첫번째: 에러정의1 -  sparse (레이블이 0~9까지 정의되어있을 때 사용)
#        에러정의2 - categorical (레이블이 원핫인코딩 되어있을 때 사용)
#optimizer = w랑 b 업데이트 / ex.그레덴트 디센트
#matrics = 모델 평가기준


#Model learning
model.fit(x_train, y_train, epochs=15) #epochs = 학습횟수
#print("model.fit: ", model.fit)

#Model verification
x_test = x_test / 255.0
print("x_test: ", x_test)
model.evaluate(x_test, y_test) #학습된 데이터가 evaluate에 자동으로 저장됨
#print("model.evaluate: ", model.evaluate)
#x_test = Feature for test
#y_test = Feature for label


```
