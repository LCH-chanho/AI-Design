```python
#1. 데이터준비
# CIFAR-10 데이터셋: 비행기, 자동차 등 10가지 종류의 이미지 6만 장으로 구성된 데이터셋
# 데이터 전처리: 이미지 픽셀 값을 0과 1 사이로 정규화하여 모델 학습에 용이하도록 만듬

#2. 모델구축
# Sequential 모델: Keras에서 제공하는 가장 간단한 모델 형태. 순차적으로 층을 쌓아 모델을 구성.
# Flatten 층: 이미지 데이터를 1차원 벡터로 펼쳐서 다음 층에 전달.
# Dense 층: 완전 연결 층으로, 모든 뉴런이 이전 층의 모든 뉴런과 연결됨.
# Activation 함수: 뉴런의 출력을 비선형적으로 변환하여 모델의 표현력을 높임.
# ReLU: 양의 입력에 대해서는 입력 값을 그대로 출력하고, 음의 입력에 대해서는 0을 출력하는 함수.
# Softmax: 다중 클래스 분류 문제에서 각 클래스에 대한 확률을 출력하는 함수.
# Compile: 모델을 컴파일하여 학습에 필요한 설정을 진행.
# optimizer: 모델의 가중치를 업데이트하는 알고리즘 (Adam, SGD 등)
# loss: 모델의 예측값과 실제 값 사이의 차이를 측정하는 함수 (sparse_categorical_crossentropy)
# metrics: 모델 성능을 평가하는 지표 (accuracy)

#3. 모델학습
# fit: 모델을 학습시킴.
# epochs: 전체 데이터셋을 몇 번 반복하여 학습할지 결정하는 파라미터
# batch_size: 한 번에 학습에 사용하는 데이터의 양
# validation_data: 모델의 일반화 성능을 평가하기 위한 검증 데이터

#4. 모델 평가
# evaluate: 테스트 데이터를 이용하여 모델의 성능을 평가






# 필요한 라이브러리 가져오기
import tensorflow as tf #딥러닝 모델 구축, 학습
from tensorflow import keras #신경망 모델 쉽게 구축
import matplotlib.pyplot as plt #데이터 시각화

# CIFAR-10 데이터 세트 로드 및 정규화
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data() #10개의 클래스로 분류된 32x32 크기의 컬러 이미지 데이터셋
X_train = X_train / 255.0 #이미지의 픽셀 값 최댓값이 255이므로 나누어서 0~1사이 값으로 만든다.
X_test = X_test / 255.0

# 모델 정의 함수 [Problem 1, 3]
def create_model():
    model = keras.Sequential([
        keras.layers.Flatten(input_shape=(32, 32, 3)), #1차원 데이터로 변경
        keras.layers.Dense(100, activation='relu'),
        keras.layers.Dropout(0.1),  # 첫 번째 은닉층에 10% 드롭아웃 적용
        keras.layers.Dense(50, activation='relu'),
        keras.layers.Dropout(0.3),  # 두 번째 은닉층에 30% 드롭아웃 적용
        keras.layers.Dense(10, activation='softmax')  # 다중 클래스 분류이므로 소프트맥스 사용
    ])
    return model


# 모델 학습 및 평가 함수
def train_and_evaluate(optimizer, batch_size, epochs=10):
    model = create_model()

     # 각 옵티마이저 인스턴스를 반복마다 새로 생성
    if optimizer_name == 'SGD':
        optimizer = keras.optimizers.SGD(learning_rate=learning_rate, momentum=0.9)
    elif optimizer_name == 'RMSprop':
        optimizer = keras.optimizers.RMSprop(learning_rate=learning_rate, rho=0.9)
    elif optimizer_name == 'Adam':
        optimizer = keras.optimizers.Adam(learning_rate=learning_rate, beta_1=0.9, beta_2=0.999)
    else:
        raise ValueError(f"Unknown optimizer: {optimizer_name}")

    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(X_train, y_train,
                        epochs=epochs,
                        batch_size=batch_size,
                        validation_data=(X_test, y_test))


    # 테스트 데이터로 평가
    test_loss, test_acc = model.evaluate(X_test, y_test)


    # 손실 함수 그래프 그리기
    plt.plot(history.history['loss'], label='train_loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.title(f"Optimizer: {optimizer_name}, Batch size: {batch_size}")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

    return test_acc

# 하이퍼파라미터 설정
learning_rate = 0.0001  # 기본 학습률

# Optimizer 이름 설정
optimizer_names = ['SGD', 'RMSprop', 'Adam']  # 각 옵티마이저 이름 리스트

# 배치 사이즈 설정
batch_sizes = [16, 32]
results = [] #결과를 results에 저장


# 6가지 케이스에 대해 학습 및 평가
for optimizer_name in optimizer_names:
    for batch_size in batch_sizes:
        test_acc = train_and_evaluate(optimizer_name, batch_size, epochs=10)
        results.append((optimizer_name, batch_size, test_acc))
        print(f"Optimizer: {optimizer_name}, Batch size: {batch_size}, Test accuracy: {test_acc}")


# 모든 결과 출력
print("\n\n\n\n")
for result in results:
    print(f"Optimizer: {result[0]}, Batch size: {result[1]}, Test accuracy: {result[2]}")
```
