```python
# 데이터 이미지 변환 후 CNN(특징추출)-DNN(분류)
# Test accuracy: 0.8352	/	소요시간: 6분42초

# 0. 라이브러리
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow import keras
import tensorflow as tf
import random
# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==


# 1. 데이터 로드
train_data = pd.read_csv("training_data.csv")
test_data = pd.read_csv("test_data.csv")

# 2. Feature와 Label 분리
X_train = train_data.iloc[:, :2].values  # In-phase, Quadrature
y_train = train_data.iloc[:, 2].values  # Label
X_test = test_data.iloc[:, :2].values
y_test = test_data.iloc[:, 2].values

# 데이터 확인: 크기 및 샘플 출력
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
print(f"First 5 y_train: {y_train[:5]}")
print(f"First 5 y_test: {y_test[:5]}")


# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==
# 1) training Data 시각화

# 데이터 읽어오기
df = pd.read_csv("training_data.csv")

# 레이블별로 데이터 분리
labels = df['Label'].unique()

# 원하는 레이블 순서 정의
desired_labels = [0, 1, 2, 3, 4, 5]

# 서브플롯 생성 (2x3)
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

# 각 레이블별 산점도 그리기
for i, label in enumerate(desired_labels):
    data = df[df['Label'] == label]
    ax = axes[i // 3, i % 3]  # 행과 열 인덱스 수정
    ax.scatter(data['In-phase'], data['Quadrature'], label=f'Label {label}', s=0.1)
    ax.set_title(f'Label {label}')
    ax.set_xlabel('In-phase')
    ax.set_ylabel('Quadrature')
    ax.grid(True)

# 전체 그림 제목 설정 (선택 사항)
plt.suptitle('training_data Distribution by Label')

# 레이아웃 조정 (선택 사항)
plt.tight_layout()

plt.show()


# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==


#가중치 초기화, 데이터 샘플링 고정
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)


# 3. 데이터 시각화와 이미지 변환 함수
def create_image_data(X, y, image_size=32, bins=32): #bins: 히스토그램의 구간 수를 설정. 값이 클수록 더 세밀한 히스토그램을 얻을 수 있음

    #2D 히스토그램을 생성하여 이미지 데이터와 레이블을 반환

    images = []
    labels = []
    unique_labels = np.unique(y) #데이터셋에 존재하는 모든 레이블(변조 방식)을 추출.

    print(f"Unique labels: {unique_labels}")  # 1) 고유 레이블 출력

    for label in unique_labels:  # 각 레이블별로 처리
        X_label = X[y == label] #각 레이블에 해당하는 데이터를 X_label에 저장

        hist, _, _ = np.histogram2d( #2차원 데이터의 히스토그램을 생성)

            X_label[:, 0], X_label[:, 1], bins=bins, range=[[-1.5, 1.5], [-1.5, 1.5]] #히스토그램의 범위를 설정
        )   #X_label[:, 0]: In-phase 값, X_label[:, 1]: Quadrature 값을 히스토그램의 x축과 y축으로 사용


        hist = hist / np.max(hist)  # 히스토그램 값을 정규화하여 0과 1 사이의 값으로 만듬 (다음 단계에서 이미지로 변환하기 위한 전처리 과정)

        # 2) 히스토그램 출력
        #print(f"Histogram for Label {label}:", hist[0])

        img_resized = np.resize(hist, (image_size, image_size))  # 크기 맞춤
        #생성된 히스토그램을 지정된 크기(image_size x image_size)의 이미지로 변환

        for _ in range(len(X_label)):  # 레이블과 동일한 개수로 반복
            images.append(img_resized) #각 레이블에 해당하는 이미지를 리스트 image에 추가
            labels.append(label) #해당 레이블을 리스트 labels에 추가
    return np.array(images), np.array(labels)


# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==


# 4. 훈련 및 테스트 데이터를 이미지로 변환
X_train_images, y_train_images = create_image_data(X_train, y_train, image_size=32)
X_test_images, y_test_images = create_image_data(X_test, y_test, image_size=32)



# CNN 입력 데이터 차원 추가
X_train_images = X_train_images[..., np.newaxis]  # (N, 32, 32, 1) 흑백이미지 채널 1
X_test_images = X_test_images[..., np.newaxis] #마지막차원에 새로운 축 추가 1

# 데이터 크기 확인
print(f"X_train_images shape: {X_train_images.shape}")
print(f"y_train_images shape: {y_train_images.shape}")
print(f"X_test_images shape: {X_test_images.shape}")
print(f"y_test_images shape: {y_test_images.shape}")


# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==
# 2) 특정 레이블에 대해 히스토그램 출력
def plot_histogram_for_label(label, X, y, bins=32):
    X_label = X[y == label]
    hist, _, _ = np.histogram2d(X_label[:, 0], X_label[:, 1], bins=bins, range=[[-1.5, 1.5], [-1.5, 1.5]])
    plt.imshow(hist.T, cmap="gray", origin="lower")
    plt.title(f"Label {label} Histogram")
    plt.colorbar()
    plt.show()

    print(f"Data for Label {label}:")
    print(X_label[:5])  # 확인하고 싶은 레이블의 데이터 샘플

# 레이블별로 히스토그램 출력
for label in [0, 1, 2, 3, 4, 5]:
    plot_histogram_for_label(label, X_train, y_train, bins=32)

# --  --  --  --  --  --  --  --  --  --  --  --  --  --  --  --  --   --  --  --  --  --  --  --  --  --  --  --

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', s=0.1)
plt.title("Data Distribution by Label")
plt.xlabel("In-phase")
plt.ylabel("Quadrature")
plt.colorbar(label="Label")
plt.show()
# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==

# 5. CNN 모델 설계
model = keras.Sequential()
model.add(keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 1)))
model.add(keras.layers.MaxPooling2D((2, 2)))
model.add(keras.layers.Conv2D(64, (3, 3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2, 2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(6, activation='softmax'))

model.summary()


# 모델 컴파일
Adam = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==

# 6. 모델 학습
history = model.fit(
    X_train_images, y_train_images,
    validation_split=0.2,  # 검증 데이터 분리
    epochs=5,
    batch_size=64
)

# 7. 모델 평가
test_loss, test_accuracy = model.evaluate(X_test_images, y_test_images)
print(f"Test accuracy: {test_accuracy:.4f}")

# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==

# 8. 학습 결과 시각화
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.legend()
plt.title("Accuracy")

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title("Loss")
plt.show()
# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==








"""
# 일반 정규화 후 간단한 DNN 사용
# Test accuracy: 0.3995 / 소요시간: 11초

# [가장 심플한 모델 + DNN]
# Test accuracy: 0.3995 (11초)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from tensorflow import keras

# 1. 데이터 로드
train_data = pd.read_csv("training_data.csv")
test_data = pd.read_csv("test_data.csv")

# 2. Feature와 Label 분리
X_train = train_data.iloc[:, :2].values  # In-phase, Quadrature
y_train = train_data.iloc[:, 2].values  # Label
X_test = test_data.iloc[:, :2].values
y_test = test_data.iloc[:, 2].values
# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==

# 3. 데이터 정규화
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)
print("정규화된 데이터 :", X_train_normalized)

# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==
# DNN 모델 구성
model = keras.Sequential()
model.add(keras.layers.Dense(64, activation='relu', input_dim=X_train_normalized.shape[1]))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(32, activation='relu'))
model.add(keras.layers.Dense(32, activation='relu'))
model.add(keras.layers.Dense(6, activation='softmax'))

# 모델 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 모델 학습
history = model.fit(X_train_normalized, y_train, epochs=20, batch_size=500, validation_split=0.2)

# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==

# 테스트 데이터 평가
test_loss, test_acc = model.evaluate(X_test_normalized, y_test)
print(f'Test accuracy: {test_acc}')

# ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==  ==
# 8. 학습 결과 시각화
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.legend()
plt.title("Accuracy")


plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title("Loss")
plt.show()

"""
```
